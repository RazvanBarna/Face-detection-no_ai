```python
from TestPy.Data_metrics.Metrics import Metrics
import pandas as pd

metrics = Metrics()
df = pd.read_csv('../CSVs/coords.csv')
print(df)
```

```python
det_boxes = {}
for _,row in df.iterrows():
    img_name = (row["filename"]).split('.')[0]
    line = [row['c_min'],row['c_max'],row['r_min'],row['r_max']]
    det_boxes.setdefault(img_name, []).append(line)

print(det_boxes)
```

```python
df2 = pd.read_csv('../CSVs/labels.csv')
gt_boxes = {}
for _,row in df2.iterrows():
    img_name = (row["filename"]).split('.')[0]
    line = [row['c_min'],row['c_max'],row['r_min'],row['r_max']]
    gt_boxes.setdefault(img_name, []).append(line)
print(gt_boxes)
```

```python
precision, recall = metrics.compute_map(det_boxes,gt_boxes,iou_threshold=0.5)
```

```python
metrics = Metrics()
df_haar = pd.read_csv('../CSVs/haar_coords.csv')
detHaar_boxes = {}
for _,row in df_haar.iterrows():
    img_name = (row["filename"]).split('.')[0]
    line = [row['c_min'],row['c_max'],row['r_min'],row['r_max']]
    detHaar_boxes.setdefault(img_name, []).append(line)

print(detHaar_boxes)
```

```python
precision, recall = metrics.compute_map(detHaar_boxes,gt_boxes,iou_threshold=0.5)
```
